## Marker color

Change the color of points (Nuqtalar rangini o'zgartiring)

Bir qancha usullari bor

1. Float number -> 1 va 0 oraliqidagi sonlarni berish
2. Colar name -> ranglar nomining bosh harfi asosan -> red r , blue b va hkz
3. Hex color code -> Color pinkerga kirib olish mumkin shakli : #4287f5
4. RGB color code -> Color pinkerga kirib olish mumkin shakli :(66/255, 135/255, 245/255)

```python
# Color namelarni ko'rish uchun

# Image(url='https://matplotlib.org/stable/_images/sphx_glr_named_colors_003_2_00x.png',height=800)

x= [3,1,-1]
y = [4,2,5]
size = [100,200, 300] # size berish
color = ['b','r','y'] # rang berish

plt.scatter(x,y, c = color, s = size)
plt.show()
```

```python
x= [3,1,-1]
y = [4,2,5]
size = [100,200, 300] # size berish
color = [(44/255, 144/255, 245/255),'#4287f5','y'] # RGB va Hex

plt.scatter(x,y, c = color, s = size)
plt.show()
```

**TITLE**

plt.title(label, loc, pad, color, fontsize)

https://matplotlib.org/stable/api/text_api.html#module-matplotlib.text text bilan ishlaydigan fnunksiyalarning deyarli hammasidan foydalanish mumkin.

**Text**

pIt.text(x, y,s, color, fontsize, fontstyle, **kwargs: text)

**xlabel and ylabel**

plt.xlabel(xlabel, labelpad, **kwargs:Text)

plt.ylabel(ylabel, labelpad, **kwargs:Text)



```python
x = [3,1,-1]
y = [4,2,5]

plt.title('DATA' ,loc='center',pad=10, fontsize=20) # loc - default center | fontsize = 10
plt.xlabel('X' ,color='red' )
plt.ylabel( 'y', rotation = 90)
plt.text(
    1,3.5,
    'mohirdev',
    size=20,
    color='#037ffc',
    weight=1000,
    style='italic'
    )

plt.scatter(x,y)
plt.show()
```

**legend**

jadval haqida malumotlarni chiqarish. **ncol** - optinal default = 1   ustunlar soni, **loc** optinal joylashuvi

pIt.legend(loc, ncol, bbox_to_anchor)

```python
x1 = [3,1,-1]
y1 = [4,2,5]

x2 = [1,1.5,0]
y2 = [3,6,2]

plt.scatter(x1,y1, label='circle')
plt.scatter(x2,y2,marker='s', label='square')

plt.legend(loc='lower right') # loc optinal
plt.show()
```

**Plot**

plt.plot()

**line** xususiyatlari:

1. color = 'blue' - rang
2. alpha = 0.5 - shaffofligi 0-1 oraliq
3. label
4. linestyle (ls) - '-' or 'solid'/ '--' or 'dashed'/ '-.' or 'dashdot'/ ':' or 'dotted'
5. linewidth(lw) - chiziq qalinligi
6. marker - burilish nuqtalari belgi
7. markeredgecolor(mec) - marker atrofi rangi
8. markeredgewidth(mew) - marker qalinligi
9. markerfacecolor(mfc) - marker rangi
10. markersize(ms) - marker o'chami

**Grid**

plt.grid() Barcha line xususiyatlaridan foydalanish mumkin

```python
x = [3,1,-1]
y = [4,2,5]

plt.plot(x, y,
         color = 'red',
         alpha=0.7,
         label='red line',
         ls = '-.',
         lw = 2,
         marker = 's',
         mec = 'blue',
         mew = 2,
         mfc = 'yellow',
         ms = 10
         ) # faqat y -> required , boshqalar optinal
plt.legend()

# plt.grid() defoult # line xususiyatlaridan foydalanish mumkin
plt.grid(axis = 'x', alpha = 0.3, color = 'blue', lw = 2, ls = ':')

plt.xticks([-1,0,1,2,3,4],['a','b','c','d','e','f']) # x o'qi malumotlarini berish mumkin
plt.yticks(range(7))
plt.show()
```

```python
y = []

for i in range(-20,21):
  y.append(i*i)

plt.title('Parbola', pad = 20, size = 20)
plt.xlabel('X', color='r')
plt.ylabel('Y', rotation = 90)
plt.plot(y,
         alpha = 0.7,
         lw = 2.5,
         ls = '--',
         color = 'b')
plt.text(16,200, 'y=x^2', size = 20, style = 'italic')
plt.grid(alpha=0.2, ls = ':')
plt.show()
```

# Bar Plot

matplotlib.pyplot.bar(x, height, width=0.8, bottom=None, *, align='center', data=None, **kwargs)

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html#matplotlib.pyplot.bar

plt.bar()

plt.barh() -> Gorizantal ko'rinish

```python
x = ['A', 'B']
h = [4,5]
w = [0.3,0.5] # alohida yoki umumiy berish mumkin
color = ['red','green']

plt.bar(x,h, width=w, color=color)
plt.show()
```

```python
data = {
    'Python': 13.33,
    'C' : 11.41,
    'C++' : 10.63,
    'Java' : 10.33,
    'C#' : 7.04
}

x = list(data.keys())
y = list(data.values())

plt.figure(figsize=(10,6))
plt.title('TIOBE index 2023', size = 20)
plt.xlabel('Programming languages', size = 15)
plt.ylabel('Ratings', size = 15)
plt.bar(x,y, color ='#1994e6')
plt.show()
```

```python
plt.barh(x,y, color ='#1994e6')
plt.show()
```

**Bar char**

```python
x = [20,20,10]

labels = ['A','B','C'] # qismlar nomi
colors = ['red','green','blue']
explodes = [0.1,0,0] # qismni ajratish

plt.pie( # faqat x required , qolganlari optinal
    x,
    labels = labels,
    colors = colors,
    explode = explodes,
    startangle = 30
)
plt.show()
```
